```python
from PIL import Image
import requests
from io import BytesIO

from torchvision import models, transforms
import torch
from tqdm import tqdm
```

```python
transform = transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
     ])
model = models.efficientnet_b0(pretrained=True)
newmodel = torch.nn.Sequential(*(list(model.children())[:-1]))
newmodel.eval()
```

```python
def extract_embedding(model, image):
    x = transform(image)
    x = x.unsqueeze(0)
    embed = model(x)
    embed = embed.squeeze()
    return embed
```

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
def get_embedding_face(model,img):
    results = model(img)
    xyxy = []
    s_max=0
    for pred in results.pred[0].tolist():
        x1,y1,x2,y2,conf,cls = pred
        x1,y1,x2,y2 = int(x1), int(y1), int(x2), int(y2)
        if int(cls)==0:
            s = abs(x2-x1)*abs(y2-y1)
            if s>s_max:
                xyxy = [x1,y1,x2,y2]
                s_max=s
    if not xyxy:
        return None
    crop_image  = img[xyxy[1]:xyxy[3], xyxy[0]:xyxy[2]]
    
    embedding = DeepFace.represent(img_path = crop_image, 
                               model_name = 'Facenet512', model = None, enforce_detection = True, 
                               detector_backend = 'retinaface', align = True, normalization = 'Facenet2018')
    
    return embedding


# import torch
# model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
# img = cv2.imread('/home/quyennt72/helloMilvus/7.jpg')
# r = get_embedding_face(model,img)
# r
```

```python
#Creat a collection
from pymilvus import CollectionSchema, FieldSchema, DataType, Collection

collection_name = "VNE_image_search_face"
dim = 1280
dim_face = 512
default_fields = [
    FieldSchema(name="id", dtype=DataType.INT64, is_primary=True, auto_id=True),
#     FieldSchema(name="norm_vector", dtype=DataType.FLOAT_VECTOR, dim=dim),
    FieldSchema(name="norm_vector_face", dtype=DataType.FLOAT_VECTOR, dim=dim_face),
]
schema = CollectionSchema(fields=default_fields, description="Collection for image in VNE, include face embedding")

collection = Collection(name=collection_name, schema=schema, consistency_level="Strong")
```

```python
# Create IVF_SQ8 index to the  collection
default_index = {"index_type": "IVF_FLAT", "params": {"nlist": 1280}, "metric_type": "IP"}
collection.create_index(field_name="norm_vector_face", index_params=default_index)
collection.load()
```

```python
random_ids = [int(red.randomkey()) for x in range(3)]
search_images = [x.decode("utf-8") for x in red.mget(random_ids)]
```

```python
search_images
```

```python
query_vectors = []
for url in search_images:
    response = requests.get(row['full_url'])
    image = Image.open(BytesIO(response.content)).convert("RGB") 
    r = extract_embedding(newmodel,image)
    r = r.tolist()
    query_vectors.append(r)
```

```python
import time
search_params = {"metric_type": "IP", "params": {"nprobe": 32}}
start = time.time()
results = collection.search(query_vectors, "norm_vector", param=search_params, limit=3, expr=None)
end = time.time() - start

print("Search took a total of: ", end)
```

```python
import matplotlib.pyplot as plt
from PIL import Image

def show_results(query, results, distances):
    
    fig_query, ax_query = plt.subplots(1,1, figsize=(5,5))
    response = requests.get(query)
    image = Image.open(BytesIO(response.content)).convert("RGB") 
    ax_query.imshow(image)
    ax_query.axis('off')
    ax_query.set_title("Searched Image")
    
    res_count = len(results)
    fig, ax = plt.subplots(1,res_count,figsize=(10,10))
    for x in range(res_count):
        response = requests.get(results[x])
        image = Image.open(BytesIO(response.content)).convert("RGB")
        ax[x].imshow(image)
        ax[x].axis('off')
        dist =  str(distances[x])
        dist = dist[0:dist.find('.')+4]
        ax[x].set_title("D: " +dist)
```

```python
for x in range(len(results)):
    query_file = search_images[x]
    result_files = [red.get(y.id).decode('utf-8') for y in results[x]]
    distances = [y.distance for y in results[x]]
    show_results(query_file, result_files, distances)
```
